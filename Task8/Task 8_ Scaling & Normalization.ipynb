```python
# Rename the 'Annual Income (k$)' and 'Spending Score (1-100)' for easy access
ma_df = ma_df.rename(columns={'Annual Income (k$)': 'Annual_Income', 'Spending Score (1-100)': 'Spending_Score'})
```

```python
# Print columns labels
ma_df.columns
```

```python
# Normalize 'Annual_Income' column manually
min_income = ma_df['Annual_Income'].min()
max_income = ma_df['Annual_Income'].max()
ma_df['Annual_Income'] = (ma_df['Annual_Income'] - min_income) / (max_income - min_income)

# Normalize 'Spending_Score' column manually
min_spending = ma_df['Spending_Score'].min()
max_spending = ma_df['Spending_Score'].max()
ma_df['Spending_Score'] = (ma_df['Spending_Score'] - min_spending) / (max_spending - min_spending)

# Display the first few rows of the dataframe with normalized columns
display(ma_df.head(10))
```

```python
# Create CSV File of Cleaned dataset
ma_df.to_csv('Mall-Customers-Cleaned.csv', index=False)
```
