```python
# Remove any columns from the DataFrame hp_df that contain more than 100 missing value (NaN).
hp_df = hp_df.drop(hp_df.isna().sum()[hp_df.isna().sum() > 100].index, axis=1)
```

```python
# Fill with mean/median/mode/domain-specific value

cat_cols = hp_df.select_dtypes(include='object').columns.tolist()
num_cols = hp_df.select_dtypes(exclude='object').columns.tolist()

# fill the catagorical columns with mode
hp_df[cat_cols] = hp_df[cat_cols].fillna(hp_df[cat_cols].mode().iloc[0])

# fill the remaining numerical columns with median
hp_df[num_cols] = hp_df[num_cols].fillna(hp_df[num_cols].median())

hp_df.head(10)
```

```python
# Identifying Outliers using IQR (Interquartile Range)
# For 'Lot Area'
Q1 = hp_df['Lot Area'].quantile(0.25)
Q3 = hp_df['Lot Area'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


print(f"\n Lot Area: Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f" Lot Area: Lower Bound (1.5*IQR): {lower_bound}, Upper Bound (1.5*IQR): {upper_bound}")
```

```python
# Print Outliers in Lot Area
outliers_lot_area = hp_df[(hp_df['Lot Area'] < lower_bound) | (hp_df['Lot Area'] > upper_bound)]

print("\nOutliers identified in 'Lot Area' using IQR:")
print(len(outliers_lot_area['Lot Area']))
```

```python
# Handling the Outliers of 'Lot Area'
hp_df['Lot Area'] = hp_df['Lot Area'].clip(lower=lower_bound, upper=upper_bound)

outliers_lot_area_cleaned = hp_df[(hp_df['Lot Area'] < lower_bound) | (hp_df['Lot Area'] > upper_bound)]
print(len(outliers_lot_area_cleaned['Lot Area']))
```

```python
# Identifying Outliers using IQR (Interquartile Range)
# For 'SalePrice'
Q1 = hp_df['SalePrice'].quantile(0.25)
Q3 = hp_df['SalePrice'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


print(f"\n SalePrice: Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f" SalePrice: Lower Bound (1.5*IQR): {lower_bound}, Upper Bound (1.5*IQR): {upper_bound}")
```

```python
# Print Outliers in  SalePrice
outliers_SalePrice = hp_df[(hp_df['SalePrice'] < lower_bound) | (hp_df['SalePrice'] > upper_bound)]

print("\nOutliers identified in 'SalePrice' using IQR:")
print(len(outliers_SalePrice['SalePrice']))
```

```python
# Handling the outliers in 'SalePrice'
hp_df['SalePrice'] = hp_df['SalePrice'].clip(lower=lower_bound, upper=upper_bound)

outliers_SalePrice_cleaned = hp_df[(hp_df['SalePrice'] < lower_bound) | (hp_df['SalePrice'] > upper_bound)]
print(len(outliers_SalePrice_cleaned['SalePrice']))

```

```python
# Handling the outliers using IQR for capping for all numerical values
num_cols = hp_df.select_dtypes(include=['float64', 'int64']).columns
Q1 = hp_df[num_cols].quantile(0.25)
Q3 = hp_df[num_cols].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


print(f"\n Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Bound (1.5*IQR): {lower_bound}, Upper Bound (1.5*IQR): {upper_bound}")
```

```python
# Cap the outliers
for col in hp_df.columns:
    if pd.api.types.is_numeric_dtype(hp_df[col]):
        hp_df[col] = hp_df[col].clip(lower=lower_bound[col], upper=upper_bound[col])
```

```python
# Create CSV File of Cleaned dataset
hp_df.to_csv('HousePrices-Cleaned.csv', index=False)
```
