Task 7: Encoding Categorical Variables

Dataset: Students Performance Dataset

```python
# Loading the dataset
st_df = pd.read_csv('/content/exams.csv')
st_df.head(10)
```

```python
# Checking Missing Values
st_df.isnull().sum()
```

```python
# Convert the 'gender' Column to Dummy variables
st_encoded = pd.get_dummies(st_df, columns=['gender'], dtype=int)
st_encoded.head(10)
```

```python
# Convert Both 'lunch, test preparation course' Columns to Dummy Variables
st_encoded = pd.get_dummies(st_encoded, columns=['lunch', 'test preparation course'], dtype=int)
st_encoded.head(7)
```

```python
st_encoded.dtypes
```

```python
# Create CSV File of Cleaned dataset
st_encoded.to_csv('Students_performance cleaned.csv', index=False)
```
