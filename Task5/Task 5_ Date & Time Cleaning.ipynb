```python
# There are 9357 valid rows of dataset and all others are null:
aq_df.dropna(how='all', inplace=True)
```

```python
aq_df.tail(5)
```

```python
aq_df.isnull().sum()
```

```python
aq_df.info()
```

```python
aq_df.shape
```

```python
# Some columns are not integer. We should change the string columns to an integer
col = ['CO(GT)', 'C6H6(GT)', 'T', 'RH', 'AH']

for i in col:
  aq_df[i] = aq_df[i].str.replace(',', '.')
  aq_df[i] = aq_df[i].astype(float)
```

```python
aq_df.head()
```

```python
aq_df.dtypes
```

```python
# Change the Hour(Time) to an integer type:
aq_df['Hour'] = aq_df['Time'].apply(lambda x: int(x.split('.')[0]))
```

```python
aq_df.dtypes
```

```python
# Convert Date to 'datetime' format
aq_df['Date'] = pd.to_datetime(aq_df['Date'], format='%d/%m/%Y')
```

```python
aq_df.head()
```

```python
aq_df.dtypes
```

```python
# Extract Year from the Date Column
aq_df['Year'] = aq_df['Date'].dt.year
```

```python
aq_df.head()
```

```python
# Extract Month and Weekday
aq_df['Month'] = aq_df['Date'].dt.month
aq_df['weekday'] = aq_df['Date'].dt.dayofweek
```

```python
# Print first Rows of dateset to check the features (Year, Month, Weekday).
aq_df.head(10)
```

```python
# Create CSV File of Cleaned dataset
aq_df.to_csv('AirQualityUCI-Cleaned.csv', index=False)
```
