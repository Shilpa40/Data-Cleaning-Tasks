```python
# Converting the 'Restaurant ID' to a string type to be treated as unique identifiers or labels
z_df['Restaurant ID'] = z_df['Restaurant ID'].astype(str)
```

```python
z_df['Restaurant ID'].dtype
```

```python
# Check for duplicates based on Restaurant ID
z_df['Restaurant ID'].duplicated().sum()
```

```python
# Check unique values in Longitude column
z_df['Longitude'].nunique()
```

```python
# Check unique values in Latitude column
z_df['Latitude'].nunique()
```

```python
# Drop Longitude and Latitude Columns, Can be helpful while using geographic maps, but we won't be doing that here and also have -ve min values
z_df.drop(['Longitude', 'Latitude'], axis=1, inplace=True)
```

```python
# Check Missing Values
z_df.isnull().sum()
```

```python
# Check number of unique cuisines
z_df['Cuisines'].nunique()
```

```python
# Check the unique cuisines names
z_df['Cuisines'].unique()
```

```python
# Filling missing values with 'No Cuisine'
z_df['Cuisines']= z_df['Cuisines'].fillna("No Cuisine")
```

```python
z_df.Cuisines.isnull().sum()
```

```python
# Identify text columns
text_columns = z_df.select_dtypes(include=['object', 'string']).columns
print("Text columns:", text_columns)
```

```python
# Check for whitespaces in text columns
for col in text_columns:
    if z_df[col].dtype == 'object' or z_df[col].dtype == 'string':
        if z_df[col].str.contains(r'^\s|\s$', na=False).any():
            print(f"Column '{col}' has leading or trailing whitespaces.")
```

```python
# Remove leading and trailing whitespaces from text columns
for col in text_columns:
    if z_df[col].dtype == 'object' or z_df[col].dtype == 'string':
        z_df[col] = z_df[col].str.strip()
```

```python
z_df.head()
```

```python
z_df['Average Cost for two'].value_counts().sort_index()
```

```python
# Calculate Q1, Q3, and IQR for 'Average Cost for two'
Q1 = z_df['Average Cost for two'].quantile(0.25)
Q3 = z_df['Average Cost for two'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = z_df[(z_df['Average Cost for two'] < lower_bound) | (z_df['Average Cost for two'] > upper_bound)]

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"\nNumber of outliers in 'Average Cost for two': {len(outliers)}")
print("\nOutliers:")
display(outliers)
```

```python
# Replace outliers with the upper bound
z_df['Average Cost for two'] = z_df['Average Cost for two'].clip(upper=upper_bound)

# Verify that the outliers have been handled by checking the count of values above the upper bound
print(f"Number of values above upper bound after handling: {len(z_df[z_df['Average Cost for two'] > upper_bound])}")
```

```python
# Calculate Q1, Q3, and IQR for 'Votes'
Q1_votes = z_df['Votes'].quantile(0.25)
Q3_votes = z_df['Votes'].quantile(0.75)
IQR_votes = Q3_votes - Q1_votes

# Define lower and upper bounds for 'Votes'
lower_bound_votes = Q1_votes - 1.5 * IQR_votes
upper_bound_votes = Q3_votes + 1.5 * IQR_votes

# Identify outliers in 'Votes'
outliers_votes = z_df[(z_df['Votes'] < lower_bound_votes) | (z_df['Votes'] > upper_bound_votes)]

print(f"Q1 for Votes: {Q1_votes}")
print(f"Q3 for Votes: {Q3_votes}")
print(f"IQR for Votes: {IQR_votes}")
print(f"Lower Bound for Votes: {lower_bound_votes}")
print(f"Upper Bound for Votes: {upper_bound_votes}")
print(f"\nNumber of outliers in 'Votes': {len(outliers_votes)}")
print("\nOutliers in Votes:")
display(outliers_votes)
```

```python
# Replace outliers with the upper bound for 'Votes'
z_df['Votes'] = z_df['Votes'].clip(upper=upper_bound_votes)

# Verify that the outliers have been handled by checking the count of values above the upper bound
print(f"Number of values above upper bound for Votes after handling: {len(z_df[z_df['Votes'] > upper_bound_votes])}")
```
